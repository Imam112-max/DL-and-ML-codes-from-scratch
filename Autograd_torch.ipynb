```python
import torch 
print(1)
```

```python
x = torch.tensor(8)#any random feature
y = torch.tensor(0) # working on the classification datasets so let's say it's flase
```

```python
w = torch.tensor(1.0 , requires_grad=True)# initializing with random weight
b = torch.tensor(0.0 , requires_grad=True)# random bias
```

```python
#defining the loss
def binary_cross_entropy(pred , actual):
    epsilon = 1e-8
    prediction = torch.clamp(pred , epsilon , 1-epsilon)
    return (actual*torch.log(prediction) + (1-actual)*torch.log(1-prediction))
```

```python
#forward bias of the nn 
z = w*x + b 
y_pred = torch.sigmoid(z)
```

```python
#now calculating the loss
loss = binary_cross_entropy(y_pred , y)
```

```python
loss.backward()
```
