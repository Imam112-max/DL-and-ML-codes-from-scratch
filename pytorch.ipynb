```python
#Tensor Operations 
#Indexing 
# Multiple Indexing  operations are performed using pytorch which are inplace changes the values of tensors without making the copy that whereas out of place operations don't 
# index_add index_add_(dim,index , tensor)
# adding along 0th axis ->columns 
x = torch.zeros(5,5)
te = torch.tensor([[1,3,5,7,9] , [1, 3 , 5,7,9] ], dtype = torch.float32)
index0 = torch.tensor([0,4])
x.index_add_(0,index0, te)
print( x)
print(torch.sum(x))
print(torch.sum(x ,dim=1))

```

```python
# Adding along 1st axis -> Rows
y = torch.zeros(2,4)
tn1 = torch.tensor([[1, 2 , 4,6] , [2, 3 , 4 ,5]] , dtype=torch.float32)
index1 = torch.tensor([0,1 ,2 ,1])
y.index_add_(1 , index1 , tn1)

```

```python
#filling along the the indices 
c = torch.randn(4,4)
index = torch.tensor([0,2])
c.index_fill_(0,index,5)
c
```
